#### 정렬, Aggregation함수, GroupBy 적용

```python
# 이름으로 정렬
titanic_sorted = df.sort_values(by=['Name'])
titanic_sorted.head(3)
```

```python
# Pclass와 Name으로 내림차순 정렬
titanic_sorted = df.sort_values(by=['Pclass', 'Name'], ascending=False)
titanic_sorted.head(3)
```

```python
df.count()
```

- groupby

```python
titanic_groupby = df.groupby('Pclass').count()
titanic_groupby
```

```python
df.groupby('Pclass')['Age'].agg([max, min])
```

```python
agg_format={'Age':'max', 'SibSp':'sum', 'Fare':'mean'}
df.groupby('Pclass').agg(agg_format)
```

- apply lambda

```python
df['Name_len']= df['Name'].apply(lambda x : len(x))
df[['Name','Name_len']].head(3)
```

```python
df['Child_Adult'] = df['Age'].apply(lambda x : 'Chlid' if x <= 15 else 'Adult')
df[['Child_Adult', 'Age']]
```
